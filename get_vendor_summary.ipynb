# Conclusion <br>
As per the above observation. We can see that data we need for the analysis is distributed across various tables. We need to create the summary table which conatin the below data:<br>
Purchase Transaction made by vendor<br>
Sales Transactions<br>
Freight cost (Cost required for the delivery of product) for each vendors<br>
Actual Product prize from the vendors. 

```python
freight_summary=pd.read_sql_query('''select "VendorNumber", sum("Freight") as Freight_cost from vendor_invoice group by "VendorNumber" ''',engine)
freight_summary
```

```python
purchase_summary=pd.read_sql_query(''' select p1."VendorNumber",p1."VendorName", p1."Brand", 
                                    avg(p1."PurchasePrice") as PurchasePrice,
                                    avg("Price") as Actual_Price,
                                    sum("Quantity") as total_quantity, 
                                    sum("Dollars") as total_dollar
                                    from purchase_prices p1 inner join purchases p2
                                    on p1."Brand" = p2."Brand"
                                    where p1."PurchasePrice">0
                                    group by p1."VendorNumber",p1."VendorName", p1."Brand" 
                                    order by total_dollar
                                    ''',engine)
purchase_summary
```

```python
sales.columns
```

```python
sales_summary= pd.read_sql_query(''' select "VendorNo","Brand", 
                                    sum("SalesQuantity") as TotalSalesQunatity,
                                    sum("SalesDollars") as TotalSalesDollars,
                                    sum("SalesPrice") as TotalSalesPrice,
                                    sum("ExciseTax") as ExciseTax
                                    from sales
                                    group by "VendorNo", "Brand" ''',engine)
sales_summary
```

```python
# Description of the brand

brand_des=pd.read_sql_query('''select "Brand","Description","Volume" from purchase_prices  ''',engine)
brand_des
```

```python
# Now we will create the final table which will contain all the date from above newly created 3 tables.

str= '''
WITH freightsummary as (
    select "VendorNumber", sum("Freight") 
    from vendor_invoice 
    group by "VendorNumber
),
purchasesummary AS (
    select p1."VendorNumber",p1."VendorName", p1."Brand", 
                                    avg(p1."PurchasePrice") as purchasePrice,
                                    avg("Price") as Actual_Price,
                                    sum("Quantity") as purchase_Qunatity, 
                                    sum("Dollars") as total_dollar
    from purchase_prices p1 inner join purchases p2
    on p1."Brand" = p2."Brand"
    where p1."PurchasePrice">0
    group by p1."VendorNumber",p1."VendorName", p1."Brand" 
    order by total_dollar
),
salessummary (
        select "VendorNo","Brand", 
                                    sum("SalesQuantity") as SalesQunatity,
                                    sum("SalesDollars") as SalesDollars,
                                    sum("SalesPrice") as SalesPrice,
                                    sum("ExciseTax") as ExciseTax
        from sales
        group by "VendorNo", "Brand"
)
select
    ps.VendorNumber, 
    ps."VendorName",
    ps."Brand",
    ps."purchasePrice",
    ps."Actual_Price",
    ps."purchase_Qunatity",
    ps."total_dollar",
    ss."SalesQunatity",
    ss."SalesDollars",
    ss."SalesPrice",
    ss."ExciseTax "
    
from purchasesummary ps join salessummary ss
on ps."VendorNumber"= ss."VendorNo"
join freightsummary fs
on ss."VendorNo"=fs."VendorNumber"
'''
Vendor_sales_summary= pd.read_sql_query(str,engine)
Vendor_sales_summary
```

```python
# We are getting few years while combining this 3 tables using the SQL. We will perform the same work by joining the dataframe that
# we have created as the summary. Those are freight_summary,purchase_summary,sales_summary.

#To join the more than two dataframes. We need the common column between them. In this case we have VendorNumber as the common column. But we need to
# rename VendorNo column of salessummar as the vendorNumber.

sales_summary.rename(columns={'VendorNo':'VendorNumber'},inplace=True)
```

```python
#Now joing the 3 dataframe

merge1=pd.merge(purchase_summary,sales_summary, on=["VendorNumber","Brand"],how='inner') #Joing based onto the more than 2 columns as the purchase & 
                                                                                        #sales summary depends onto the vendornumber & brand. 
final=pd.merge(merge1,freight_summary,on="VendorNumber",how='inner')
vendor_summary=pd.merge(final,brand_des,on="Brand",how='inner')

vendor_summary
```

```python
# Now we will write this final table into our database

final.to_sql("vendor_summary", engine,if_exists='replace', index=False )
```

## Creating a New Columns for our further Analysis. 

```python
# Gross Profit which shows the profit gained by that vendor.It is calulated by difference between sales prize & Purchase Prize.

vendor_summary["GrossProfit"]= vendor_summary['totalsalesdollars']- vendor_summary['total_dollar']

# display(vendor_summary[vendor_summary["Gross Profit"] <=0]) 
# Here gross profit negative indicates the loss. 
```

```python
# Profit Margin shows the percenatge of the profit earned. 
vendor_summary["ProfitMargin"]= (vendor_summary["GrossProfit"] / vendor_summary['totalsalesdollars']) * 100
```

```python
# Stock Turnover 1 means vendor has sold all the product which was purchased. It means sales qunatity & Purchase Qunatity is same. 
# If it is more than 1 means. There was already present stock at the vendor side.

vendor_summary["StockTurnover"]= vendor_summary['totalsalesqunatity'] / vendor_summary['total_quantity']
```

```python
#SalesPurchaseRatio is the ratio of the total sales prize to total purchase Prize.

vendor_summary["SalesPurchaseRatio"]= vendor_summary['totalsalesdollars'] / vendor_summary['total_dollar']
```
